```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from evidently.report import Report
from evidently.metric_preset import DataDriftPreset
```

---

## 🌸 Section 2: Load Dataset (Iris)

```python
data = load_iris(as_frame=True)
df = data.frame

# Rename columns for clarity
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target']

# Simulate current production and baseline
baseline = df.sample(frac=0.5, random_state=42).reset_index(drop=True)
current = df.sample(frac=0.5, random_state=99).reset_index(drop=True)
```

---

## 🧪 Section 3: Inject Synthetic Drift

```python
# Add Gaussian noise to simulate feature drift
current_drifted = current.copy()
current_drifted['petal_width'] += np.random.normal(loc=0.5, scale=0.3, size=len(current))

print("Drift injected on: petal_width")
```

---

## 📊 Section 4: Generate Evidently Report

```python
report = Report(metrics=[DataDriftPreset()])
report.run(reference_data=baseline, current_data=current_drifted)
report.show(mode='inline')  # or mode='dash' to launch dashboard
```
